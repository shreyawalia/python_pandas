```python
#we can use string method in dataframe
#to know language worked with we can directly query like
df['LanguageWorkedWith']
```

```python
#to have only those rows that have python in it we can use string to find such rows as
filt = df['LanguageWorkedWith'].str.contains('Python', na = False)
#be careful with Na as presence of Na gives error ValueError: cannot mask with array containing NA / NaN values
```

```python
df.loc[filt, 'LanguageWorkedWith']
```

```python
#filt gives us true false values ans so when we apply it to our loc it masks out the false values.
```
