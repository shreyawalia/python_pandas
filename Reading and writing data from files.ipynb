```python
import pandas as pd
```

```python
#reading data with csv
```

```python
df = pd.read_csv("data/survey_results_public.csv")
df
```

```python
#filtering to get only columns with india
filt = df['Country'] == 'India'
india = df.loc[filt]
india
```

```python
india.set_index('Respondent', inplace = True)
```

```python
india.head()
```

```python
#exporting as csv
#creates new csv file at the location specified
india.to_csv('data/modified.csv')
```

```python
#tab delimited file(tsv) #like csv but separatedf by tab
#exporting and creating tsv
india.to_csv('data/modified.tsv',sep = '\t')
```

```python
#reading tsv file
pd.read_csv('data/modified.tsv',sep = '\t')
```

```python
#Excel file
#pip install packages xlwt xlrd openpyxl
#writing to excel file
#advanced args can be passed to write to a particular sheet or column
import openpyxl
india.to_excel('data/modified.xlsx')
```

```python
#reading from excel file
test = pd.read_excel('data/modified.xlsx', index_col = 'Respondent')
test
```

```python
#json 
#writing to json
#simple writing to json produces results that are different than expected it is dictionary like by default
#to orient it in different way(make it list like) we need to pass args
#line = true for new line
india.to_json('data/modified.json', orient = 'records', lines = True)
```

```python
#reading from json 
test = pd.read_json('data/modified.json', orient = 'records', lines = True)
```
